```python
import pandas as pd
import os
```

```python
csv_file = "../files/calcite_stmts.tsv"  
df = pd.read_csv(csv_file, sep="\t")
```

```python
df.head()
```

```python
len(df)
```

```python
# check duplicates
duplicates = df.duplicated(subset=['stmt_app_name', 'stmt_id'], keep=False)

if duplicates.any():
    print("there are duplicated rows")
    print(df[duplicates])
else:
    print("there is no duplicated rows")
```

```python
for _, row in df.iterrows():
    filename = f"../query/{row['stmt_id']}.sql"
    
    with open(filename, "w", encoding="utf-8") as f:
        f.write(row["stmt_raw_sql"])
    
    print(f"Saved: {filename}")
```
